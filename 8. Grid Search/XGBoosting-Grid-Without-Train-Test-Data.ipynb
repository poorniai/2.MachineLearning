```python
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'max_depth': [3, 5, 7]
  
}


grid=GridSearchCV(GradientBoostingRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input = float(input("Enter Age:::"))
bmi_input = float(input("Enter bmi:::"))
childern_input = float(input("Childern:::"))
sex_male_input = int(input("Sex Male 0 or  1:"))
smoker_yes_input = int(input("Smoker Yes 0 or 1"))
```

```python
Future_Predictions=grid.predict([[age_input,bmi_input,childern_input,sex_male_input,smoker_yes_input]])
Future_Predictions
```
