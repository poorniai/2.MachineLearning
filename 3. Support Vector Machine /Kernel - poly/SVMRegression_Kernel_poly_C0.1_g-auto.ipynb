```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
```

```python
independent
```

```python
dependent=dataset[['Profit']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=.30,random_state=0)
X_train
y_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly',C=0.1,gamma='auto')
result=regressor.fit(X_train,y_train)
result
```

```python
result
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
r_score
```
